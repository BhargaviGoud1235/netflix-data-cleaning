# Netflix Data Cleaning Project 🍿
## Objective
Clean and preprocess the Netflix Movies and TV Shows dataset using Pandas.

We will handle missing values, fix data types, clean messy columns, and export a clean dataset.

```python
# Import Libraries
import pandas as pd
import numpy as np
```

## 1. Load Dataset

```python
df = pd.read_csv('../data/netflix_titles.csv')
df.head()
```

## 2. Dataset Overview

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

## 3. Check Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

## 4. Handle Missing Values

```python
# Fill categorical missing values
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Not Available')
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

## 5. Convert Date Column

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'].head()
```

## 6. Clean Duration Column

```python
# Extract numeric duration
df['duration_number'] = df['duration'].str.extract(r'(\d+)').astype(float)

# Extract unit (min or Seasons)
df['duration_unit'] = df['duration'].str.extract('([A-Za-z]+)')

df[['duration', 'duration_number', 'duration_unit']].head()
```

## 7. Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

## 8. Clean Column Names

```python
df.columns = df.columns.str.lower()
df.columns
```

## 9. Final Check

```python


df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])

df['duration'] = df['duration'].fillna("Unknown")
df['duration_number'] = df['duration_number'].fillna(0)
df['duration_unit'] = df['duration_unit'].fillna("Unknown")
```

```python
df.info()
df.isnull().sum()
```

# 10. FINAL VALIDATION CHECK

```python

print("Missing values:\n", df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
```

# 11. Save cleaned dataset

```python
df.to_csv('../data/netflix_titles_cleaned.csv', index=False)
print('Cleaned dataset saved successfully!')
```

- Fixing data types
